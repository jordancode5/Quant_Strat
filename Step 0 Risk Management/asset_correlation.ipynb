```python
# download necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Get your Correlation matrices

```python
# Step 1: Download price data
tickers = ["NVDA", "SPY", "QQQ", "BRK-B", "AAPL", "MSFT", "TSLA", "AMZN", "GOOG", "UUUU", "BABA", "JPM", "SOXL"]
data = yf.download(tickers, start="2022-01-01", end="2025-08-22")["Close"]

# Step 2: Calculate daily returns
returns = data.pct_change().dropna()

# Step 3: Correlation matrix (for visualization)
corr_matrix = returns.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Asset Correlation Matrix")
plt.show()

# Step 4: Covariance matrix (for optimization)
cov_matrix = returns.cov()

print("Correlation matrix:\n", corr_matrix)
print("\nCovariance matrix:\n", cov_matrix)
```
