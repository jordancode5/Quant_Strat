For the first example I will be using a dice roll


```python
import random           # main package for random number generation
import matplotlib.pyplot as plt # package for plotting
import numpy as np
```

First do a single die to confirm behavior and practice using 

```python
def roll_one_die(num_rolls):
    """Simulate rolling one six-sided dice a specified number of times.

    Args:
        num_rolls (int): The number of times to roll the dice.

    Returns:
        list: A list containing the value of single die rolled.
    """
    results = []
    for _ in range(num_rolls):
        die1 = random.randint(1, 6)  # Roll first die
        # die2 = random.randint(1, 6)  # Roll second die
        results.append(die1)   # Sum the results
    return results

def plot_results(results, dice_count):
    # Create histogram with 6 bins
    
    num_bins = 6 + (dice_count - 1) * 5

    edges = np.arange(dice_count - 0.5, 6 * dice_count + 1.5, 1)
    # bins=num_bins

    plt.hist(results, bins = edges, edgecolor='black')
    # plt.xticks(range(dice_count, 6*dice_count + 1))  # Set x-ticks to show all possible sums

    # Add labels and title
    plt.xlabel("Value Range")
    plt.ylabel("Frequency")
    plt.title("Histogram of Value Ranges for Rolling {} Dice".format(dice_count))

    # Show plot
    plt.show()
```

```python
# plot the results
num_rolls = 10000
dice_count = 1
results = roll_one_die(num_rolls)
plot_results(results, dice_count)
```

The data above is more or less random (take note. This process is not truly random). Now let's do this simulation with multiple dice

Make a function that will do this with any number of dice you choose

```python
def roll_mult_die(num_rolls, dice_count):
    """Simulate rolling one six-sided dice a specified number of times.

    Args:
        num_rolls (int): The number of times to roll the dice.

    Returns:
        list: A list containing the value of single die rolled.
    """
    results = []
    dice_sum = 0        # initialize sum of dice
    for _ in range(num_rolls):
        dice_sum = 0
        for die in range(dice_count):
            dice_sum += random.randint(1, 6)    # Roll each die and add to sum

        results.append(dice_sum)   # Sum the results
    return results
```

Let's do this with 1, 2, 3, 4, 5, 6, 9, and 15 dice
You should notice the distribution getting tighter and tighter

```python
num_rolls = 10000
dice_count = 1
dice_counts = [1, 2, 3, 4, 5, 6, 9, 20]
for dice_num in dice_counts:
    results = roll_mult_die(num_rolls, dice_num)
    plot_results(results, dice_num)
```
